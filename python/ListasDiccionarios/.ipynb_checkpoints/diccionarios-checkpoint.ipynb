# Diccionarios

Colección de elementos donde cada valor está asociado a una clave única.  

Las propiedades de los diccionarios son:
- Son dinámicos, ya que pueden crecer o decrecer, ya que se pueden añadir o eliminar elementos
- Son indexados los elementos del diccionario, ya que son accesibles a través del key
- Son anidados, ya que un diccionario puede contener otro diccionario en el campo `value`

Estos son generados con las llaves `{}` y para cada elemento se define la estructura `clave:valor`, de forma equivalente se puede crear utilizando el método `dict()` e introduciendo los pares `clave:valor`

```python
colores = {'amarillo':'yellow','azul':'blue'}
colores
```

```python
datos = dict([
    ('Nombre', 'Sara'),
      ('Edad', 27),
      ('Documento', 1003882),
])
datos
```

## Métodos

Los diccionarios tienen varios métodos integrados que nos permiten manipular sus datos de forma eficiente. Al igual que con las listas estos métodos nos ayudan añadir, eliminar o consultar información.  

### Métodos de consulta y acceso

- `get(clave)` Devuelve el valor de una clave, si la clave no existe, devuelve `None` en lugar de un error
- `keys()` Devuelve una lista con todas las claves del diccionario
- `values()` Devuelve una lista con todos los valores
- `items()` Devuelve una lista de tuplas con cada par `clave:valor`. Es ideal para recorrer diccionarios con bucles

### Métodos de modificación

- `update(otro_diccionario)` Actualiza el diccionario con los pares clave-valor de otro o añade los nuevos
- `setdefault(clave, valor_defecto)` Devuelve el valor si la clave existe, si no, la crea con el valor por defecto que le des

### Métodos de eliminación

- `pop(clave)` Elimina la clave y te devuelve el valor
- `popitem()` Elimina y devuelve el último par clave-valor añadido
- `clear()` Vacía el diccionario por completo, dejandolo vacío `{}`
