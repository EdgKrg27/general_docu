## Tipos de datos

Definen que clase de información estamos manejando. Existen varios tipos de datos en python y en todos los lenguajes de programación, lo más utilizados son:

- Enteros (int): Números sin decimales, como 5 o -10
- Flotantes (float): Números con decimales como 3.14, 2.0
- Cadenas (str): Texto encerrado en comillas, como "Hola Mundo"
- Booleanos (bool): VAlores de lógica que solo pueden ser True o False

En python se conoce como tipo de dato dinámico, a diferencia de otros lenguajes de programación donde una variable nace con un tipo y muere con ese mismo tipo, en Python las variables son nombres que aputan a *"objetos"* de memoria.

```python
print(type(5)) # entero
print(type(256745674567.5234353463546)) # flotante
print(type('Hola Mundo')) # string o cadena de texto
print(type(True)) # bool
```

Se pueden realizar operaciones con estos datos, pero también definen que tipo de operaciones son permitidas. Aunque es dinámico por los objetos, Python es fuertemente tipado, es decir, no se pude mezclar tipos de datos no compatibles

```python
# esta operación si se puede, a pesar que son dos tipos de datos diferentes, son compatibles
print(5 + 7.3) # int + float
```

```python
# esta operación no se permite ya que los tipos de datos no son compatibles
print("7" + 3) # str + int
```

ya que python se necesita ser explisito con las operaciones, para solucionar el problema de querer sumar un string con un número cual sea, se necesita hacer la converisón de tipo de dato del string, cabe mencionar que solamente la conversión de tipo de dato funcionara si este es posible en caso contrario retornara un error de conversión.

Para transformar datos, usamos funciones que tienen el mismo nombre que el tipo de dato:

- `str()` convierte a texto
- `int()` convierte a entero
- `float()` convierte a flotante

```python
print(int("7") + 3)
```

## Variables

Las variables son como etiquetas que le pones a un valor para poder identificarlo y usarlo más tarde. A diferencia de otro lenguajes, no necesitas declarar que tipo de dato vas a guardar, python lo deduce automáticamente, a esto se le conoce como tipo dinámico.

```python
# asignación de variables
a = 7
b = "3"
c = True
```

Otro benficio que se tiene en Python son la asignación multiple de variables, esto hace que el código sea mucho más limpio y legible.

Existen dos formas principales de hacerlo:

1. Mismo valor para todas: Si se quiere que varias variables empiecen con el mismo dato
2. Valores disitntos (desempaquetado): Se puede asiganar valores diferentes a variables respetando el orden.

```python
# Opción 1
a =  b = 7
print(a)
print(b)
```

```python
# Opción 2
a, b = 10, "Edgar"
print(a)
print(b)
```
