# Ejercicios de diccionarios

## El Escenario
Eres el responsable de la gestion de un almacén. Cada producto se identifica como una clave única pero su información es un conjuto de datos fijos.

```python
# Definición de la estructura
# Usaremos namedtuple para que el código sea legible y un diccionario para el inventario
from collections import namedtuple

# Definición de la estructura fija del producto: (ID, Nombre,Categoría)
producto = namedtuple("Producto", ["id", "nombre", "categoria"])

# Diccionario: Clave = código de barrar (int), valor = producto (tupla), stock (int), precio (float)
inventario = {
    1001: [producto(1, "Laptop", "Electrónica"), 5, 1200.50],
    1002: [producto(2, "Teclado", "Periféricos"), 15, 45.00],
    1003: [producto(3, "Monitor", "Electrónica"), 8, 300.00]
}

# Operacion de búsqueda y actualización
def realizar_venta(codigo, cantidad):
    if codigo in inventario:
        # extraemos los datos (desempaquetado de la lista valor)
        info_id, stock, precio = inventario[codigo]

        if stock >= cantidad:
            # Actualizamos el stock en el diccionario
            inventario[codigo][1] -= cantidad
            total = cantidad * precio
            print(f"Venta exitosa: {cantidad} x {info_id.nombre}")
            print(f"Total a cobrar: {total}")
        else:
            print(f"Stock insuficiente: solo quedan {stock} unidades.")
    else:
        print("Código del producto no existe")
```

```python
realizar_venta(1002, 2)
```

```python
# genear reporte: debe recorrer el diccionaro e imprimir: Producto [nombre] | Categoría [Categoría] 
# | valor total en stock: $[stock * precio]
def generar_reporte(dic):
    for clave, valor in dic.items():
        print(f"Producto: {valor[0].nombre} | Categoría: {valor[0].categoria} | Valor total: {valor[1] * valor[2]}")
```

```python
generar_reporte(inventario)
```
