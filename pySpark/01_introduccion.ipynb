```python
# creacioń del spark context
sc = spark.sparkContext
```

```python
# Creación de un RDD vacio
rdd_vacio = sc.emptyRDD()
rdd_vacio.collect()
```

```python
# Creación de RDD vacio con tres particiones
rdd_vacio_3particiones = sc.parallelize([], 3)
rdd_vacio_3particiones.getNumPartitions()
```

```python
# creación de un RDD a partir de una lista
rdd = sc.parallelize([1,2,3,4,5,6,7,8,9])
rdd.collect()
```

```python
# Creación de un RDD a partir de un archivo de texto
rdd_texto = sc.textFile('./data.txt')
rdd_texto.collect()
```

```python
# Creación de un RDD a partir de una operación
rdd_suma = rdd.map(lambda x: x**2)
rdd_suma.collect()
```

```python
# Creacioń de un DataFrame
df = spark.createDataFrame([(1,'Jose'), (2,'Juan')], ['id', 'nombre'])
df.show()
```

```python
# Convertir DataFrame a RDD
rdd_df = df.rdd
rdd_df.collect()
```

```python
# Ejercicios
# Cree una sesión de Spark con nombre Cap2 y asegúrese de que emplea todos los cores disponibles para ejecutar en su ambiente de trabajo
spark = SparkSession.builder.appName("Cap2").master("local[*]").getOrCreate()
sc = spark.sparkContext
```

```python
# Cree dos RDD vacíos, uno de ellos no debe contener particiones y el otro debe tener 5 particiones. Utilice vías diferentes para crear cada RDD
rdd_vacio_1 = sc.emptyRDD()
rdd_vacio_2 = sc.emptyRDD().partitionBy(5)
```

```python
print(rdd_vacio_1.collect())
print(rdd_vacio_2.collect())
print(rdd_vacio_2.getNumPartitions())
```

```python
# cree un RDD que contenga los números primos que hay entre 1 y 20
rdd_primos = sc.parallelize([2,5,7,11,13,17,19])
rdd_primos.collect()
```

```python
# Cree un nuevo RDD a partir del RDD creado en el ejercicio anterior el cuál solo contenga los números primos mayores a 10
rdd_filter = rdd_primos.filter(lambda x : x > 10)
rdd_filter.collect()
```
